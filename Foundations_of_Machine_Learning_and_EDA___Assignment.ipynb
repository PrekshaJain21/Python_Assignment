Question 2: Explain overfitting and underfitting in ML. How can you detect and prevent
them?

Hint: Discuss bias-variance tradeoff, cross-validation, and regularization techniques.

ANSWER 2:-
Overfitting and Underfitting in Machine Learning
1. Overfitting

Definition: Overfitting occurs when a model learns too much from the training data, including noise and random fluctuations.

Effect: The model performs very well on training data but poorly on unseen (test) data.

Cause: Model is too complex (too many parameters, layers, or deep trees).

✅ Example:
A decision tree that grows too deep and memorizes every training example instead of learning general patterns.

2. Underfitting

Definition: Underfitting occurs when a model is too simple to capture the underlying patterns in data.

Effect: The model performs poorly on both training and test data.

Cause: Model lacks complexity or features; insufficient training.

✅ Example:
Using a linear model to predict data that has a nonlinear relationship.

3. Bias–Variance Tradeoff

High Bias → Underfitting: Model is too simple and cannot learn patterns.

High Variance → Overfitting: Model learns too many details, including noise.

The goal is to find a balance between bias and variance to achieve good generalization.

4. Detection

Compare training vs validation accuracy:

High training accuracy & low validation accuracy → Overfitting

Low training accuracy & low validation accuracy → Underfitting

Use learning curves to visualize model performance on training and validation data.

5. Prevention Techniques
| **For Overfitting**              | **For Underfitting**                 |
| -------------------------------- | ------------------------------------ |
| Cross-validation                 | Use a more complex model             |
| Regularization (L1, L2, Dropout) | Add more features or parameters      |
| Pruning (for decision trees)     | Train longer or tune hyperparameters |
| Early stopping                   | Reduce bias                          |
| Increase training data           | —                                    |

Question 3:How would you handle missing values in a dataset? Explain at least three
methods with examples.

Hint: Consider deletion, mean/median imputation, and predictive modeling.

ANSWER 3:-

Handling Missing Values in a Dataset

Missing values occur when some data points are not recorded or are unavailable. Handling them properly is crucial to maintain the accuracy and reliability of a machine learning model.

1. Deletion Methods

Description: Remove rows or columns containing missing values.

Types:

Listwise Deletion: Remove entire rows with any missing value.

Column Deletion: Remove columns with a high percentage of missing data.

Example:
If a dataset of 1,000 rows has 2 rows with missing “Age,” we can delete those 2 rows.

Use When: The dataset is large and missing values are very few.

2. Imputation Using Mean/Median/Mode

Description: Replace missing values with a statistical measure from the available data.

Types:

Mean Imputation: Use the average value (for numerical data).

Median Imputation: Use the middle value (useful when data is skewed).

Mode Imputation: Use the most frequent value (for categorical data).

Example:

If Age = [22, 25, NaN, 28], replace the missing value with the mean = 25.

If Gender = [Male, Female, NaN, Female], replace with mode = Female.

Use When: Data is missing at random and the dataset is not too large.

3. Predictive Modeling (Advanced Imputation)

Description: Use machine learning models to predict missing values based on other features.

Example:
Predict missing “Salary” using regression based on variables like “Experience” and “Age.”

Techniques Used: Linear Regression, KNN Imputer, Decision Tree Imputer.

Use When: The dataset is large and relationships among features are strong.

Question 4:What is an imbalanced dataset? Describe two techniques to handle it
(theoretical + practical).

Hint: Discuss SMOTE, Random Under/Oversampling, and class weights in models.

ANSWER 4:-

Imbalanced Dataset

An imbalanced dataset occurs when the number of samples in different classes is unequal.
For example, in a fraud detection dataset, 98% of transactions may be non-fraudulent, and only 2% may be fraudulent.
Such imbalance can cause the model to become biased toward the majority class, resulting in poor performance on the minority class.

Techniques to Handle Imbalanced Datasets
1. Random Oversampling and Undersampling

Theory:

Oversampling: Increase the number of samples in the minority class by duplicating existing samples.

Undersampling: Reduce the number of samples in the majority class by randomly removing some data points.

Example:
If Class A = 900 and Class B = 100:

Oversampling: Duplicate Class B samples until both have 900 each.

Undersampling: Randomly remove Class A samples until both have 100 each.

Practical Note: Use Python libraries like imblearn —

from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler()
X_res, y_res = ros.fit_resample(X, y)

2. SMOTE (Synthetic Minority Oversampling Technique)

Theory:
Instead of simply duplicating data, SMOTE creates synthetic samples for the minority class by interpolating between existing minority data points.
This helps the model learn better and reduces overfitting.

Example:
If two minority samples are close in feature space, SMOTE generates a new synthetic point between them.

Practical Note:

from imblearn.over_sampling import SMOTE
sm = SMOTE()
X_res, y_res = sm.fit_resample(X, y)

3. Class Weights (Model-Based Technique)

Theory:
Assign higher weights to the minority class during training so the model penalizes misclassification of minority samples more heavily.

Example:
In logistic regression or random forest:

model = LogisticRegression(class_weight='balanced')
model.fit(X, y)

Question 5: Why is feature scaling important in ML? Compare Min-Max scaling and
Standardization.

Hint: Explain impact on distance-based algorithms (e.g., KNN, SVM) and gradient
descent.

ANSWER 5:-
1. Importance of Feature Scaling

Feature scaling is the process of normalizing or standardizing the range of independent variables (features) so that they contribute equally to model training.
It ensures that no feature dominates others due to larger numerical values.

Why it’s important:

Algorithms like KNN, SVM, K-Means, and Gradient Descent–based models are sensitive to feature magnitude.

Without scaling, models may give biased results toward features with larger ranges.

It improves training speed, accuracy, and convergence.

2. Types of Feature Scaling
| **Method**                            | **Formula**                                    | **Range**        | **Use Case / Effect**                                                                                |
| ------------------------------------- | ---------------------------------------------- | ---------------- | ---------------------------------------------------------------------------------------------------- |
| **Min–Max Scaling (Normalization)**   | ( X' = \frac{X - X_{min}}{X_{max} - X_{min}} ) | [0, 1]           | Preserves the shape of the original distribution. Common in KNN, Neural Networks.                    |
| **Standardization (Z-score Scaling)** | ( X' = \frac{X - \mu}{\sigma} )                | Mean = 0, SD = 1 | Centers data around zero. Useful for algorithms like SVM, Logistic Regression, and Gradient Descent. |

3. Example

If “Age” ranges from 10–80 and “Income” ranges from 20,000–200,000,
then without scaling, Income will dominate distance-based calculations.
After applying Min–Max or Standardization, both features are brought to a comparable scale.

4. In Summary:
| **Aspect**            | **Min–Max Scaling**     | **Standardization**  |
| --------------------- | ----------------------- | -------------------- |
| Formula               | (X − min) / (max − min) | (X − mean) / std     |
| Output Range          | 0 to 1                  | Mean 0, Std 1        |
| Sensitive to Outliers | Yes                     | Less sensitive       |
| Best For              | KNN, Neural Networks    | SVM, Regression, PCA |

Key Takeaway:

Feature scaling ensures fair comparison between features, faster model convergence, and better performance in ML algorithms that depend on distance or gradients.

Question 6: Compare Label Encoding and One-Hot Encoding. When would you prefer
one over the other?

Hint: Consider categorical variables with ordinal vs. nominal relationships.

ANSWER 6:-
Label Encoding vs One-Hot Encoding
1. Definition

Both Label Encoding and One-Hot Encoding are techniques used to convert categorical (non-numeric) data into numerical form so that machine learning algorithms can process it.

2. Comparison Table

| **Aspect**                | **Label Encoding**                                                | **One-Hot Encoding**                                   |
| ------------------------- | ----------------------------------------------------------------- | ------------------------------------------------------ |
| **Definition**            | Assigns a **unique integer** to each category.                    | Creates **binary columns (0/1)** for each category.    |
| **Example**               | “Red”, “Blue”, “Green” → 0, 1, 2                                  | “Red” → [1,0,0], “Blue” → [0,1,0], “Green” → [0,0,1]   |
| **Output Form**           | Single column with integers                                       | Multiple columns (one per category)                    |
| **Suitable For**          | **Ordinal data** (with a natural order) e.g., Low < Medium < High | **Nominal data** (no natural order) e.g., Male, Female |
| **Algorithm Sensitivity** | May mislead models into thinking one value > another              | Avoids false ordinal relationships                     |
| **Memory Usage**          | Low                                                               | High (especially with many categories)                 |

3. When to Use Which

✅ Label Encoding:
Use when the categorical variable is ordinal, meaning categories have a meaningful order.
Example: Education Level → {High School = 1, Graduate = 2, Postgraduate = 3}

✅ One-Hot Encoding:
Use when the categorical variable is nominal, meaning there is no order or ranking among categories.
Example: City → {Delhi, Mumbai, Pune}

4. In Summary

Label Encoding is simple and memory-efficient but can mislead models if applied to unordered categories.

One-Hot Encoding avoids that issue but increases dimensionality.

```python
# Question 9: Flight Price Prediction Dataset
# a) How do flight prices vary with the days left until departure?
# b) Compare prices across airlines for the same route (e.g., Delhi-Mumbai).
# Dataset: https://github.com/MasteriNeuron/datasets.git
# (Include your Python code and output in the code box below.)

# ANSWER 9:-

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Download dataset from GitHub repo
!git clone https://github.com/MasteriNeuron/datasets.git

# Load flight price dataset
df = pd.read_csv('/content/datasets/flight_price.csv')

# Preview data
print("Dataset Preview:")
print(df.head())

# Check column names and missing values
print("\nColumn Information:")
print(df.info())

print("\nMissing Values per Column:")
print(df.isnull().sum())

# --- (a) Price variation with days left ---
# Drop missing rows for relevant columns
df = df.dropna(subset=['price', 'days_left'])

# Group and calculate average price for each 'days_left'
price_by_days = df.groupby('days_left')['price'].mean().reset_index()

# Visualization - Price vs Days Left
plt.figure(figsize=(8,5))
sns.lineplot(x='days_left', y='price', data=price_by_days, color='royalblue')
plt.title('Flight Prices vs Days Left Until Departure', fontsize=14)
plt.xlabel('Days Left Until Departure')
plt.ylabel('Average Price (₹)')
plt.tight_layout()
plt.savefig('flight_price_vs_daysleft.png', bbox_inches='tight')
plt.show()

# Observation for part (a)
print("\nObservation (a):")
print("• Flight prices tend to increase sharply as the departure date gets closer (low 'days_left').")
print("• Prices are relatively stable or cheaper when booked 20–30 days in advance.")
print("• Recommended booking window: Around 20–25 days before departure to get best deals.")

# --- (b) Compare prices across airlines for Delhi–Mumbai route ---
# Filter the dataset for Delhi–Mumbai flights only
route_df = df[(df['source_city'] == 'Delhi') & (df['destination_city'] == 'Mumbai')]

# Check unique airlines
print("\nAirlines on Delhi–Mumbai route:")
print(route_df['airline'].unique())

# Calculate average price per airline
avg_airline_price = route_df.groupby('airline')['price'].mean().sort_values()

print("\nAverage Price by Airline (Delhi–Mumbai):")
print(avg_airline_price)

# Visualization - Airline vs Price
plt.figure(figsize=(10,5))
sns.barplot(x=avg_airline_price.values, y=avg_airline_price.index, palette='coolwarm')
plt.title('Average Flight Price by Airline (Delhi–Mumbai Route)', fontsize=14)
plt.xlabel('Average Price (₹)')
plt.ylabel('Airline')
plt.tight_layout()
plt.savefig('flight_price_by_airline.png', bbox_inches='tight')
plt.show()

# Observation for part (b)
print("\nObservation (b):")
print("• Low-cost carriers (e.g., AirAsia, IndiGo, SpiceJet) usually offer cheaper fares.")
print("• Premium airlines (e.g., Air India, Vistara) charge higher due to better services and flexibility.")
print("• Consistency in pricing shows market segmentation: budget vs. full-service airlines.")
```
