Q1. What is the difference between a function and a method in Python?
Function: A block of code that performs a specific task. It is defined using the def keyword and can be called directly.

Method: A function that is associated with an object or class. It is always called using the object it belongs to.

Key Differences:

Aspect	Function	Method
Definition	Independent block of code	Function that belongs to an object
Calling	Called directly using its name	Called using object (e.g., obj.method())
Association	Not tied to any object	Always tied to an object/class

Example:

# Function
def greet(name):
    return f"Hello, {name}!"

print(greet("Ritesh"))  

# Output: Hello, Ritesh!

# Method
text = "hello world"
print(text.upper())  

# Output: HELLO WORLD

Q5. What are iterators in Python and how do they differ from iterables?

Iterable → Any object that can return its elements one by one (list, tuple, string).

Iterator → An object created from an iterable using iter(), which produces elements one by one using next().

Key Differences:

Aspect	Iterable	Iterator
Definition	Collection of elements	Produces elements one by one
Created by	Default collections	iter(iterable)
Access	Loops (for)	next()
Exhaustible	No	Yes

Example:

fruits = ["apple", "banana", "cherry"]  
# Iterable

fruits_iter = iter(fruits)

 # Iterator

print(next(fruits_iter))  

#output: apple

print(next(fruits_iter))  

#output: banana

Q6. Explain the concept of generators in Python and how they are defined.

Generator = Special type of iterator.

Defined using a function with yield OR using generator expressions.

Generates values on demand (lazy evaluation).

Example using yield:

def greet_generator(names):
    for name in names:
        yield f"Hello {name}!"

gen = greet_generator(["Ritesh", "Alex", "Maya"])
print(next(gen))

 #output: Hello Ritesh!

print(next(gen))  

#output: Hello Alex!


Example using generator expression:

squares = (x**2 for x in range(3))

print(next(squares))  

#output: 0

print(next(squares))  

#output: 1

```python
# Q1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_of_even(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum
my_list = [1, 2, 3, 4, 5, 6]
print("Sum of even numbers:", sum_of_even(my_list))
```

```python
# Q4. Write a Python function that checks if a given number is prime or not and then print all prime numbers from 1 to 200.
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
primes = [num for num in range(1, 201) if is_prime(num)]
print("Prime numbers from 1 to 200:", primes)
```

```python
# Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms
        self.count = 0
        self.a, self.b = 0, 1
    def __iter__(self):
        return self
    def __next__(self):
        if self.count < self.n_terms:
            if self.count == 0:
                self.count += 1
                return 0
            elif self.count == 1:
                self.count += 1
                return 1
            else:
                self.a, self.b = self.b, self.a + self.b
                self.count += 1
                return self.b
        else:
            raise StopIteration
fib = FibonacciIterator(10)
print("Fibonacci sequence:", list(fib))
```

```python
# Q6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i
print("Powers of 2 up to exponent 5:")
for value in powers_of_two(5):
    print(value)
```

```python
# Q7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(filename):
    with open(filename, "r") as file:
        for line in file:
            yield line.strip()
with open("sample.txt", "w") as f:
    f.write("Hello World\n")
    f.write("Python Generators\n")
    f.write("Read file line by line\n")
print("File content:")
for line in read_file_line_by_line("sample.txt"):
    print(line)
```

```python
# Q9. Write a Python program that uses map() to convert a list of temperatures from Celsius to Fahrenheit.
def c_to_f(celsius):
    return (celsius * 9/5) + 32
celsius_temps = [0, 20, 37, 100]
fahrenheit_temps = list(map(c_to_f, celsius_temps))
print("Celsius temperatures:", celsius_temps)
print("Fahrenheit temperatures:", fahrenheit_temps)
```

```python
# Q10. Create a Python program that uses filter() to remove all the vowels from a given string.
def remove_vowels(s):
    vowels = "aeiouAEIOU"
    return "".join(filter(lambda ch: ch not in vowels, s))
text = "Hello Python Programming"
result = remove_vowels(text)
print("Original string:", text)
print("String without vowels:", result)
```

```python
# Q11. Book shop accounting routine using lambda and map
# Given data: [Order Number, Book Title, Quantity, Price per Item]
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einführung in Python3, Bernd Klein", 3, 24.99]
]
print(orders)
# Lambda function to calculate total with condition
calc_total = lambda order: (
    order[0],  # Order Number
    order[2] * order[3] if order[2] * order[3] >= 100 else order[2] * order[3] + 10
)
# Using map()
result = list(map(calc_total, orders))
# Example usage
print("Final order totals:", result)
```

```python
# Q11. Write a Python program, which returns a list with 2–tuples.Each tuple consists of the order number and the product of the price per item and the quantity. The product should be increased by 10 € if the value of the order is smaller than 100 €.
# Sample data: [Order Number, Book Title, Quantity, Price per Item]
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einführung in Python3, Bernd Klein", 3, 24.99]
]
result = []
for order in orders:
    order_number = order[0]
    total_price = order[2] * order[3]
    if total_price < 100:
        total_price += 10
    result.append((order_number, round(total_price, 2)))
print("Final order totals:", result)
```
